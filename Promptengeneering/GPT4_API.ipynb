```python
import openai

# Set up your OpenAI API key
openai.api_key = 'your-api-key'

def analyze_sentiment_korean(text):
    prompt = f"""
    <instruction>
    Write a complete answer to the positive and negative sentiment analysis for the given Korean sentences. When writing your answer, you must use the following parameters: sentiment = classify as positive/negative, probability = find the probability value with three decimal places between 1 and 0 with 0 being negative and 1 being positive, extract positive_topics, extract negative_topics.
    Please print the output in Korean
    </instruction>
    <Question>
    {text}
    </Question>
    """
    
    response = openai.Completion.create(
        engine="gpt-4",
        prompt=prompt,
        max_tokens=150,
        temperature=0.7,
    )
    
    return response.choices[0].text.strip()

# Example sentences
sentences = [
    "이번 스마트폰의 디자인은 마음에 드는데 성능은 별로에요. 배터리 수명이 너무 짧고 카메라 화질도 떨어져요.",
    "넘나 이쁘고 발색 발림성 다 좋아요 잘 산 거 같아요",
    "이 앱은 사용하기 어려워요. 인터페이스가 직관적이지 않습니다."
]

results = [analyze_sentiment_korean(sentence) for sentence in sentences]
for result in results:
    print(result)
```
