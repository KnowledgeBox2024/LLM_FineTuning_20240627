Generate Python source code that reflects a template with optimized prompt engeneering based on Clauade3 sonet for a wide variety of sentences and reliable analysis results.


```python
import anthropic
import os

# Initialize the Anthropic client
client = anthropic.Anthropic(api_key=os.environ.get("ANTHROPIC_API_KEY"))

def analyze_sentiment(sentence):
    prompt = f"""You are an expert in Korean language sentiment analysis. Analyze the following Korean sentence and provide a detailed sentiment analysis.

Sentence: "{sentence}"

Please provide the analysis in the following format:
sentiment: [positive/negative/neutral]
probability: [probability value with three decimal places between 0 and 1, where 0 is very negative and 1 is very positive]
positive_topics: [list of positive topics or aspects mentioned]
negative_topics: [list of negative topics or aspects mentioned]

Ensure your analysis is thorough and accurate, considering the nuances of the Korean language and cultural context.

분석 결과를 한국어로 출력해주세요."""

    response = client.messages.create(
        model="claude-3-sonnet-20240229",
        max_tokens=300,
        temperature=0,
        messages=[
            {"role": "user", "content": prompt}
        ]
    )
    
    return response.content

# Example usage
sentences = [
    "이번 스마트폰의 디자인은 마음에 드는데 성능은 별로에요. 배터리 수명이 너무 짧고 카메라 화질도 떨어져요.",
    "넘나 이쁘고 발색 발림성 다 좋아요 잘 산 거 같아요",
    "이 앱은 사용하기 어려워요. 인터페이스가 직관적이지 않습니다.",
    # Add more sentences here for testing
]

for sentence in sentences:
    print(f"Sentence: {sentence}")
    analysis = analyze_sentiment(sentence)
    print(f"Analysis:\n{analysis}\n")
```
